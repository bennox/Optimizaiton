### Functions for the simulation of the models

```python
from numpy.random import multivariate_normal
from scipy.linalg.special_matrices import toeplitz
from numpy.random import randn


def simu_linreg(coefs, n_samples=1000, corr=0.5):
    """Simulation of a linear regression model
    
    Parameters
    ----------
    coefs : `numpy.array`, shape=(n_features,)
        Coefficients of the model
    
    n_samples : `int`, default=1000
        Number of samples to simulate
    
    corr : `float`, default=0.5
        Correlation of the features

    Returns
    -------
    A : `numpy.ndarray`, shape=(n_samples, n_features)
        Simulated features matrix. It samples of a centered Gaussian 
        vector with covariance given by the Toeplitz matrix
    
    b : `numpy.array`, shape=(n_samples,)
        Simulated labels
    """
    # Construction of a covariance matrix
    cov = toeplitz(corr ** np.arange(0, n_features))
    # Simulation of features
    A = multivariate_normal(np.zeros(n_features), cov, size=n_samples)
    # Simulation of the labels
    b = A.dot(coefs) + randn(n_samples)
    return A, b
```

## Proximal operators and Solver


We remind that the proximal operator of a fonction $g$ is given by:

$$
\text{prox}_g(y, t) = \arg\min_x \Big\{ \frac 12 \|x - y\|_2^2 + t g(x) \Big\}.
$$

where $t \geq 0$ is a non-negative number.
We have in mind to use the following cases

- Lasso penalization, where $g(x) = s \|x\|_1$
- Indicator function of $\mathbb{R}^n_+$, where $g(x) = i_{x \in \mathbb{R}_+^n}(\cdot)$

where $s \geq 0$ is a regularization parameter.

We want to minimize:
$$
\arg\min_x F(x)
$$
with
$$
 F(x) = \frac{1}{2} \|b - Ax\|^2 + g(x)
$$

```python
#Computation of g(x) for lasso penalization 

def prox_lasso(x, s, t=1.):
    """Proximal operator for the lasso at x with strenght t"""
    return np.sign(x) * np.maximum(np.abs(x)- s*t, 0)

def lasso (x, s):
    return s*np.sum(abs(x))

#Computation of g(x) for indicator function

def indicator (x, s=0., t=1.):
    if min(x) <0:
        return np.inf
    else:
        return 0
    
def prox_indicator(x,s=0. ,t=1.): # We set s and t parameters so the prox function has the same form as the lasso one.
    return x.clip(min = 0) #We project all negative coefficients on R+. 
```

### Question 2: Justify why proximal coordinate descent can be applied to obtain a minimum of such objective functions.

The first term of $F$ is quadrative thus convex and differentiable. The second term $g(x)$ is separable in $g_i$, which are convex. Indeed, for the lasso penalization, the absolute value is convex and for the second problem the indicator function is also convex. The conditions for proximal coordinate descent are then satisfied.

### Question 3 : implementing proximal coordinate descent

#### Implementing "smart" residuals updates

At each iteration, we need to compute : $x_{k+1}^i = prox_{\gamma_i,g_i}(x_k^{(i)}-\gamma_i \nabla_i f(x_k) ) $ (the update the $i^th$ coordinate. Since, we can write : $\nabla_i f(x_k) = A_i^T (Ax_k-y)$. Therefore, at each coordinate we need to compute the whole matrix product which seems "stupid" (in term of speed and memory).

Let $r_k = Ax_k-y$ the residuals. We can notice that $r_{k+1} = r_k + (x_k^{(i)}-x_{k-1}^{(i)})A_i $. Thus, we don't have do calculate the whole matrix at each iteration but only one component.

#### Proximal coordinate descent algorithm

```python
from numpy import linalg

def cd_linreg(x0, A, b, g, prox_g, s=0., n_iter=50,
              x_true=coefs, verbose=True):
    """Proximal gradient descent algorithm

    Minimize :
    
    1/2 ||b−Ax||^2 + s * g(x)
    
    with coordinate descent.
    """
    x = x0.copy()
    x_new = x0.copy()
    n_samples, n_features = A.shape

    # estimation error history
    errors = []
    # objective history
    objectives = []
    # Current estimation error
    err = linalg.norm(x - x_true) / linalg.norm(x_true)
    errors.append(err)
    # Current objective
    obj = 0.5 * linalg.norm(b - A.dot(x))**2 + g(x, s)
    objectives.append(obj)

    
#    if verbose:
#print("Launching Coordinate Descent solver...")
#print(' | '.join([name.center(8) for name in ["it", "obj", "err"]]))

    L = np.sum(A * A, axis = 0)
    residuals = np.dot(A,x) - b
    
    for k in range(n_iter + 1):

        #### TODO ####
        i = k % n_features
        x_new[i] = prox_g(x_new[i] - (1/L[i])*np.dot(A[:,i].T,residuals),s=s)
        residuals = residuals + (x_new[i] - x[i])*A[:,i]
        x = x_new.copy()
        
        obj = 0.5 * linalg.norm(b - A.dot(x))**2 + g(x, s)
        err = linalg.norm(x - x_true) / linalg.norm(x_true)
        errors.append(err)
        objectives.append(obj)
#        if k % 10 == 0 and verbose:
#            print(' | '.join([("%d" % k).rjust(8), 
#                              ("%.2e" % obj).rjust(8), 
#                              ("%.2e" % err).rjust(8)]))
    return x, objectives, errors
```
